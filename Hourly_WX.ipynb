```python
import pandas as pd
import numpy as np
from functools import reduce
from datetime import timezone, datetime,date
import pytz

pac = pytz.timezone('America/Los_Angeles')
utc = pytz.timezone('UTC')
```

```python
def Kel_To_Far(k):
    #c = k + 273
    #f = ((9 * c) / 5) + 32
    f = 1.8*(k - 273) + 32
    return f
```

```python
path = './Data/historical-hourly-weather-data/temperature.csv'
temp = pd.read_csv(path,index_col='datetime',parse_dates = [0])
Drop_col = [col for col in temp.columns if 'Seattle' not in col]
temp.drop(Drop_col,axis=1,inplace=True)
temp.rename(columns={'Seattle': 'Temp_k'},inplace=True)
temp['Temp_F'] = temp.apply(Kel_To_Far)
#temp.sample(15)
```

```python
path = './Data/historical-hourly-weather-data/weather_description.csv'
weather_cond = pd.read_csv(path,index_col='datetime',parse_dates = [0])
Drop_col = [col for col in weather_cond.columns if 'Seattle' not in col]
weather_cond.drop(Drop_col,axis=1,inplace=True)
weather_cond.rename(columns={'Seattle': 'weather_cond'},inplace=True)
#weather_cond.sample(15)
```

```python
path = './Data/historical-hourly-weather-data/pressure.csv'
pressure = pd.read_csv(path,index_col='datetime',parse_dates = [0])
Drop_col = [col for col in pressure.columns if 'Seattle' not in col]
pressure.drop(Drop_col,axis=1,inplace=True)
pressure.rename(columns={'Seattle': 'pressure'},inplace=True)
#pressure.sample(15)
#pressure.index.dtype
```

```python
path = './Data/historical-hourly-weather-data/wind_speed.csv'
wind = pd.read_csv(path,index_col='datetime',parse_dates = [0])
Drop_col = [col for col in wind.columns if 'Seattle' not in col]
wind.drop(Drop_col,axis=1,inplace=True)
wind.rename(columns={'Seattle': 'wind_speed'},inplace=True)
#wind.sample(15)
```

```python
dfs = (temp,weather_cond,pressure,wind)
WX = reduce(lambda left,right: pd.merge(left,right,on = 'datetime'),dfs)


#localize as UTC then convert to Pacific
WX = WX.tz_localize(utc,ambiguous = True, nonexistent = 'shift_forward').tz_convert(pac)

WX.drop('Temp_k',axis=1,inplace=True)

WX.sort_index(inplace = True,ascending=False)

WX.index.names=['Date_Join']

#WX = pd.get_dummies(data=WX, columns = ['weather_cond'])
#WX.sample(10)
```

```python
fre = './Data/Bike_Counts/Fremont_Bridge.csv'
fremont = pd.read_csv(fre,parse_dates = [0],index_col='Date')

fremont.rename(columns = {'Fremont Bridge East Sidewalk' : 'EastSide',
                         'Fremont Bridge West Sidewalk':'WestSide',
                         'Date':'Date_Join'} , inplace= True)


fremont = fremont.tz_localize(pac,ambiguous = True, nonexistent = 'shift_forward')
fremont['Mean_Xing'] = fremont[['EastSide','WestSide']].mean(axis=1)

fremont.index.names=['Date_Join']
```

```python
data = fremont.merge(WX, how = 'inner',left_on = 'Date_Join',right_on = 'Date_Join')
data['Hour'] = data.index.strftime('%H')
```

```python
null_df = pd.DataFrame(data.isnull().sum().sort_values(ascending=False),columns=['Null_Count'])
null_df[null_df['Null_Count'] >0]
#data[data.isnull().any(axis=1)]
```

```python
data['pressure'].fillna(value = data['pressure'].mean(),inplace=True)
data['EastSide'].fillna(value = data['EastSide'].mean(),inplace=True)
data['WestSide'].fillna(value = data['WestSide'].mean(),inplace=True)
data['Mean_Xing'].fillna(value = data['Mean_Xing'].mean(),inplace=True)
data['Temp_F'].fillna(value = data['Temp_F'].mean(),inplace=True)
```
